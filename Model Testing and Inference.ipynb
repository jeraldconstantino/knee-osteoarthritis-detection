### I. PROBLEM STATEMENT

<p align="justify">&emsp;&emsp;The World Health Organization (2021) reported that approximately 343 million individuals worldwide suffer from osteoarthritis [1], and it is a major cause of disability among adults aging over 60 years old [2]. Knee Osteoarthritis (KOA) is a chronic condition with progressive degeneration of articular cartilage, leading to bone friction, stiffness, and a painful stress fracture. It is often diagnosed manually based on the patient’s symptoms, history, laboratory results, and imaging [3]. This approach is technician-dependent, labor-intensive, and time-consuming. Radiologist sometimes recognizes an imprecise result. Thus, the researchers proposed an innovative way of KOA detection using the Deep Learning inference approach. The researchers trained and evaluated the AI model using the Knee Osteoarthritis Dataset downloaded from Kaggle. The AI model is designed to classify whether the patient has osteoarthritis (distinguished when there is articular cartilage loss) or a healthy knee. <br>

<br>
<b>Reference:</b><br>
<b>[1]</b> Z. Wang, T. Efferth, X. Hua, and X. Zhang, “Medicinal plants and their secondary metabolites in alleviating knee osteoarthritis: A systematic review," Phytomedicine, 105, 154347, 2022. doi:10.1016/J.PHYMED.2022.154347 <br>
<b>[2]</b> S. L. Kolasinski et al., “2019 American College of Rheumatology/Arthritis Foundation Guideline for the Management of Osteoarthritis of the Hand, Hip, and Knee,” Arthritis Care and Research, 72(2), 149–162, 2020. doi:10.1002/ACR.24131/ABSTRACT <br>
<b>[3]</b> P. A. Sukerkar, and Z. Doyle, “Imaging of Osteoarthritis of the Knee,” Radiologic Clinics of North America, 60(4), 605–616, 2022. doi:10.1016/J.RCL.2022.03.004</p>

### II. DATASET DESCRIPTION

<p align="justify">&emsp;&emsp;The Knee Osteoarthritis Dataset with Severity Grading, organized by the Osteoarthritis Initiative (OAI), comprised knee X-ray image data to detect knee joints and knee KL grading. The dataset has five classifications with an unbalanced number of images available: <br></p>

* <b>Grade 0</b> (Healthy) contains <b>3,253</b> images in total.
* <b>Grade 1</b> (Doubtful) contains <b>1,495</b> images in total.
* <b>Grade 2</b> (Minimal) contains <b>2,175</b> images in total.
* <b>Grade 3</b> (Moderate) contains <b>1,086</b> images in total.
* <b>Grade 4</b> (Severe) contains <b>251</b> images in total.<br>

<p align="justify">&emsp;&emsp;However, this study is only focused on two classifications for KOA detection: Healthy or Degenerative. Another class is only to detect the knee bone. Three hundred knee X-ray images for each class are used for training and testing the AI model. The researchers divided the image dataset for each class (healthy and degenerative) following the 80:20 ratio. 80% of the data is used for training the AI model, while 20% is used for testing purposes. </p>

<b>Note:</b>&emsp;The dataset is available [here](https://www.kaggle.com/datasets/shashwatwork/knee-osteoarthritis-dataset-with-severity?select=auto_test).
